Definiendo g1 


$x = \frac{5}{2x-1}$

```python
import matplotlib.pyplot as plt
```

```python
def g2(x):
    return 5/(2*x-1)
```

```python
x_vec=[]
x_vec.append(0)
contador = 0
error_vec=[]
error_vec.append(1)
criterio_paro = 0.0001
```

```python
while(error_vec[contador] > criterio_paro):
    x_temp = g2(x_vec[contador])
    x_vec.append(x_temp)
    e_temp = abs(x_vec[contador+1]-x_vec[contador])
    error_vec.append(e_temp)
    contador += 1
```

```python
print(x_vec)
```

```python
print(error_vec)
```

```python
plt.plot(x_vec)
plt.show()
```

```python
plt.plot(error_vec)
plt.show()
```
